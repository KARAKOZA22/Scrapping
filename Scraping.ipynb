```python
import pandas as pandas
import requests
from bs4 import BeautifulSoup
```

## Book name, Book URL, Number of pages, Number of downloads

```python
url='https://www.pdfdrive.com/machine-learning-books.html'
page = requests.get(url)
soup = BeautifulSoup(page.content,"html.parser")
soup
```

```python
book_name = []
book_url = []
no_of_pages = []
no_of_downloads = []

for z in range(2,60):
    BookName = soup.find_all(class_="ai-search")
    BookUrl = soup.find_all('div',class_="file-right")
    NoOfPages = soup.find_all(class_="fi-pagecount")
    NoOfDownloads = soup.find_all(class_="fi-hit")
    for i in BookUrl:
        x = 'https://www.pdfdrive.com'
        s=i.find("a").get("href")
        book_url.append(x+s)

    for i in BookName:
        book_name.append(i.text.strip())

    for i in NoOfPages:
        no_of_pages.append(i.text.strip())

    for i in NoOfDownloads:
        no_of_downloads.append(i.text.strip())
    w = f"https://www.pdfdrive.com/search?q=machine%20learning&pagecount=&pubyear=&searchin=&page={z}"
    
```

```python
df = pd.DataFrame(list(zip(book_name,book_url,no_of_pages,no_of_downloads)),
               columns =['book name', 'book url','no of pages','no of downloads'])
```
